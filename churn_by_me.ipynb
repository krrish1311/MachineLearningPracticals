```python
import pandas as pad
import numpy as num
```

```python
dataset=pad.read_csv("Churn_Modelling.csv")
```

```python
dataset.head(40)
```

```python
dataset.columns
```

```python
dataset.corr()
```

```python
X=dataset[['CreditScore', 'Geography',
       'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard',
       'IsActiveMember', 'EstimatedSalary']]
```

```python
y=dataset["Exited"]
```

```python
geography=dataset["Geography"]
geography=pad.get_dummies(geography,drop_first=True)
```

```python
gender = dataset["Gender"]
gender=pad.get_dummies(gender,drop_first=True)
```

```python
tenure=dataset["Tenure"]
tenure=pad.get_dummies(tenure,drop_first=True)
```

```python
products=dataset["NumOfProducts"]
products=pad.get_dummies(products,drop_first=True)
```

```python
active=dataset["IsActiveMember"]
active=pad.get_dummies(active,drop_first=True)
```

```python
dataset.columns
```

```python
X=dataset[['CreditScore','Age', 'Balance','EstimatedSalary']]
```

```python
hascrcard=dataset["HasCrCard"]
hascrcard=pad.get_dummies(hascrcard,drop_first=True)
```

```python
X_final=pad.concat([X,gender,geography,tenure,products,hascrcard,active],axis=1)
```

```python
X_final
```

```python
from keras.models import Sequential
```

```python
mind=Sequential()
```

```python
mind.get_config
```

```python
from keras.layers import Dense
```

```python
mind.add(Dense(bias_initializer="zeros",kernel_initializer="zeros",activation="relu",units=12,input_dim=22))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.33, random_state=42)
```

```python
mind.get_config()
```

```python
mind.summary()
```

```python
mind.add(Dense(activation="relu",units=8))
```

```python
mind.add(Dense(kernel_initializer="zeros",bias_initializer="zeros",activation="relu",units=6))
```

```python
mind.get_config()
```

```python
mind.summary()
```

```python
mind.add(Dense(activation="sigmoid",units=1))
```

```python
mind.get_config()
```

```python
from keras.optimizers import Adam 
```

```python
mind.compile(optimizer=Adam(learning_rate=0.01),loss="binary_crossentropy")
```

```python
mind.fit(X_train,y_train,epochs=100)
```

```python
mind.predict(X_test)
```
